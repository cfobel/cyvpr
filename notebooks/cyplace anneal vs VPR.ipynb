Since the `cyplace`-annealer _(i.e., `cyplace.bin.run_anneal`)_ uses an
anneal-schedule ported from the VPR annealing schedule, we would expect the
placement results to be similar _(ideally with no statistical difference)_ to
those produced by VPR.

This notebook aims to investigate whether this is the case, and if not, what
might be the problem(s).

## Plot `VPR`-anneal vs. `cyplace`-anneal behaviour

[Back to top](#Table-of-contents) $\cdot$
[Parent](#Compare-clma-results)

Below, we plot how the following stats trend throughout the `cyplace`-anneal
process.  Each curve represents the _mean_ value of each of several trials,
which only differ by starting `seed` value.  The error bars designate the
range of each value at the corresponding outer-loop iteration across the
trials.

```python
def plot_h5fs(h5fs, net_file_namebase, xlim_=None, ylim_=None):
    stats_tables = {}
    stats_types = {}
    max_iterations = 0
    for label, h5f in h5fs.iteritems():
        title('[%s] %s' % (label, net_file_namebase))
        stats_frame = get_placement_stats_frame(h5f,
                                                net_file_namebase)
        if 'mean_cost' in stats_frame:
            # Assume VPR, so use `vpr_placement_stats_tables`
            stats_tables[label] = vpr_placement_stats_tables(
                    stats_frame)
            stats_types[label] = ('temperature', 'mean_cost',
                                  'radius_limit', 'success_ratio')
        else:
            # Assume `cyplace`, so use
            # `cyplace_placement_stats_tables`
            stats_tables[label] = cyplace_placement_stats_tables(
                    stats_frame)
            stats_types[label] = ('temperature', 'cost',
                                  'radius_limit', 'success_ratio')
        max_iterations = max(max_iterations,
                             len(stats_tables[label].values()[0]))
        plot_stats_tables(stats_tables[label],
                          label_prefix='[%s] ' % label,
                          stats=stats_types[label])
        if xlim_ is None:
            xlim(0, max_iterations)
        else:
            xlim(*xlim_)
        if ylim_ is None:
            ylim(0, 1.4)
        else:
            ylim(*ylim_)
        figure()
```

```python
from collections import OrderedDict

h5fs = OrderedDict([
        ('VPR fast', ts.open_file('../../synced_results/placed-'
                                 'results-fast-bounding_box.h5')),
        ('cyplace [radius-limit disabled]', ts.open_file('../../synced_results/placed-results-cyplace-51cb06cb.h5')),
        ('cyplace [radius-limit zero-allowed]', ts.open_file('../../dev-cyplace/placed-clma-s14-0a932cadc8937e837c39e6541d5f54b6c28cc1b8.h5')),
        ('cyplace [radius-limit min-clamped to 1]', ts.open_file('../../synced_results/placed-results-cyplace-7a8d71d4.h5')),
])
```

In the following sections, we plot the results for placement several MCNC netlists, using
`VPR` and `cyplace`.  For `VPR`, all trials are run using the _bounding-box_ algorithm, with `fast` enabled.  For `cyplace`, we plot several variations of the annealing algorithm, with different handling of the radius-limit.  We discuss the `cyplace` variations in detail below.

### VPR, _bounding-box_ with `fast` enabled

As shown below, in the `VPR anneal`, once the success-ratio drops to ~40%,
changes in radius-limit maintain the success-ratio at ~40% until about the 110th outer-loop iteration.

```python
plot_h5fs(dict(h5fs.items()[:1]), net_file_namebase)
```

### `cyplace` with `radius_limit` _disabled_

Due to an oversight, the annealing behaviour in `cyplace` revision `51cb06cb` 
does not respect the _radius-limit_ of the current schedule when selecting
a random target position for each swap.  In other words, during each swap, a
block may potentially move to any other position in the grid, rather than
being constrained to within a distance defined by the _radius-limit_.  Below,
we can see that this has a dramatic impact on the behaviour of the anneal.

Notice that the success-ratio of the `cyplace`-anneal seems to be a
monotonic-decreasing function, whereas in the `VPR anneal`, once the
success-ratio drops to ~40%, changes in radius-limit maintain the
success-ratio at ~40% until about the 110th outer-loop iteration.

The programming oversight was discovered since there appeared to be no effect
on the rate-of-change of the success-ratio even during a significant drop-off
in _radius-limit_, which was unexpected.  The fact that the `cyplace`-anneal
did not match the behaviour of the `VPR fast`-anneal was surprising, since
the trends seem very similar until the radius-limit starts to drop.  This lead us to check whether the radius-limit was actually being honoured in the `cyplace`-anneal.

```python
plot_h5fs(dict(h5fs.items()[1:2]), net_file_namebase,
          xlim_=(0, 140))
```

### `cyplace` with `radius_limit` allowed to reach zero

Based on the discovery that the _radius-limit_ was not being handled in
revision `51cb06cb` of the `cyplace` anneal, we modified the code to use the
_schedule's_ current _radius-limit_ value to constrain the potential target
positions of each swap.  As shown below, this dramatically improves the
behaviour of the `cyplace`-anneal, such that the trends match quite well with
the behaviour of `VPR`.  However, note that there is some sporadic
fluctuation between iterations 120 and 140.  We conjectured that these spikes
in success-rate were due to the _radius-limit_ being allowed to take on a
value of zero.  The impact of _radius-limit_ value of zero is the _only_
target position that would be considered for each block during a swap would
be the block's current position.  Since the anneal unconditionally accepts
any move that is either improving or does not change cost, _all_ swaps would
be accepted with a _radius-limit_ of zero.  __N.B.__, the success-ratio is
slightly less than one, since the radius-limit was only activated for the
logic-blocks _(i.e., not input-output blocks)_.

```python
plot_h5fs(dict(h5fs.items()[2:3]), net_file_namebase,
          xlim_=(0, 140))
```

### `cyplace` with `radius_limit` minimum-clamped to one

Based on the results from the previous section, we modified the anneal to
keep the _radius-limit_ from dropping below a value of one.  This was done by
applying a minimum-clamping of the value to one, such that any value of the
_radius-limit_ in the schedule below one would simply result in selecting
target positions a single tile away in the $x$ and $y$ dimensions.

By employing the minimum-clamping to one, the sporadic behaviour from the
previous section was eliminated, as shown below.

```python
plot_h5fs(dict(h5fs.items()[3:]), net_file_namebase,
          xlim_=(0, 140))
```

## How a target position is selected

Specifically, given a _source-slot_, $s$, a _radius-limit_,
$r$, and a grid with origin, $g$, and extents, $e$, select a
_target-slot_ as follows:

1. Look up the tile-coordinates of the source slot, $x(s), y(s)$.
1. Randomly choose a target $x$-coordinate, $x(t)$, such that:

  * $max(x(s) - r, x(g)) \le x(t) \lt min(x(s) + r, x(g) + x(e))$
1. Repeat step __2__ for $y(t)$.
