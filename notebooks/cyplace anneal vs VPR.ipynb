# Overview

[Back to top](#Table-of-contents) $\cdot$
[Next](#Compare-clma-results)

Since the `cyplace`-annealer _(i.e., `cyplace.bin.run_anneal`)_ uses an
anneal-schedule ported from the VPR annealing schedule, we would expect the
placement results to be similar _(ideally with no statistical difference)_ to
those produced by VPR.

This notebook aims to investigate whether this is the case, and if not, what
might be the problem(s).

```python
import tables as ts

from cyvpr.bin.place_results_summary import (
        get_placement_stats_frame, cyplace_placement_stats_tables,
        vpr_placement_stats_tables, plot_stats_tables)

# Make tables size nicely to half a full HD monitor.
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

# Compare `clma` results

[Back to top](#Table-of-contents) $\cdot$
[Previous](#Overview) $\cdot$

Let's start by looking at the placement results for `clma`, using `cyplace`,
then using `VPR` _(in `fast`-mode)_.

```python
net_file_namebase = 'clma'
h5fs = {}
stats_tables = {}
```

## Generate `cyplace` stats pivot-table

[Back to top](#Table-of-contents) $\cdot$
[Parent](#Compare-clma-results) $\cdot$
[Next](#Generate-VPR-fast-stats-pivot-table)

Below, we plot how the following stats trend throughout the `cyplace`-anneal
process.  Each curve represents the _mean_ value of each of several trials,
which only differ by starting `seed` value.  The error bars designate the
range of each value at the corresponding outer-loop iteration across the
trials.

```python
label = 'cyplace'
h5fs[label] = ts.open_file('../../synced_results/'
                           'placed-results-cyplace-51cb06cb.h5')
h5f = h5fs[label]
stats_frame = get_placement_stats_frame(h5f, net_file_namebase)
# Use `cyplace_placement_stats_tables` since we are processing
# `cyplace` results.
stats_tables[label] = cyplace_placement_stats_tables(stats_frame)
```

```python
label = 'VPR fast'
h5fs[label] = ts.open_file('../../synced_results/'
                           'placed-results-fast-bounding_box.h5')
h5f = h5fs[label]
stats_frame = get_placement_stats_frame(h5f, net_file_namebase)
# Use `vpr_placement_stats_tables` since we are processing VPR
# results.
stats_tables[label] = vpr_placement_stats_tables(stats_frame)
```

## Plot `cyplace` results and `VPR fast` results

[Back to top](#Table-of-contents) $\cdot$
[Parent](#Compare-clma-results) $\cdot$
[Previous](#Generate-VPR-fast-stats-pivot-table)

Below we plot the results for placement of the netlist in VPR net-file,
`clma`, using `cyplace` and `VPR fast`.

Notice that the success-ratio of the `cyplace`-anneal seems to be a
monotonic-decreasing function, whereas in the `VPR anneal`, once the
success-ratio drops to ~40%, changes in radius-limit maintain the
success-ratio at ~40% until about the 110th outer-loop iteration.

__NB__ __TODO__: We need to figure out why the success-ratio in the
`cyplace`-anneal does not match the behaviour of the `VPR fast`-anneal.  This
seems surprising, since the trends seem very similar until the radius-limit
starts to drop.  Things to check:

  * Is the radius-limit actually being honoured in the `cyplace`-anneal?
    * This should be checked, since there appears to be no effect on
      the rate-of-change of the success-ratio even during a significant
      drop-off in _radius-limit_, which is unexpected.

```python
plot_stats_tables(stats_tables['cyplace'],
                  label_prefix='[cyplace] ',
                  stats=('temperature', 'cost', 'radius_limit',
                         'success_ratio'))
figure()
plot_stats_tables(stats_tables['VPR fast'],
                  label_prefix='[VPR fast] ',
                  stats=('temperature', 'mean_cost', 'radius_limit',
                         'success_ratio'))
```
