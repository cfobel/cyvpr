In this section, we discuss the post-routing results of placing all 20 MCNC
netlist benchmarks using:

 1. VPR's place bounding-box annealer,
    running in _fast_ mode _(i.e., `inner_num=1`)_
 2.  VPR's bounding-box place annealer,
    running in default _fast_ mode _(i.e., `inner_num=10`)_
 3. `cyplace` annealer using _Star+_
 
For each annealer, we conducted 10 trials, _i.e._, we placed each netlist 10
times with 10 unique random-number-generator seeds.

```python
def flatten_dict_to_dataframe(data_frames_by_label,
                              label_name='label'):
    label_len = max([len(k) for k in data_frames_by_label.keys()])
    data_array = np.array([(label, ) + tuple(d)
                           for label, df in data_frames_by_label
                           .iteritems()
                           for field, d in df.iterrows()],
                          dtype=zip([label_name, ]
                                    + list(df.keys()),
                                    ('S%d' % label_len, )
                                    + tuple(df.dtypes)))
    return pd.DataFrame(data_array)
```

```python
def get_flattened_route_states(h5fs, label_name='label'):
    data_frames = OrderedDict([(label,
                        OrderedDict([(net_file_group._v_name,
                                       get_routing_data_frame(
                                        net_file_group.route_states))
                                     for net_file_group in h5f.root]))
                       for label, h5f in h5fs.iteritems()])
    
    data = OrderedDict([(label,
                        OrderedDict([(net_file_group._v_name,
                                      min_success_data(data_frames
                                                       [label]
                                                       [net_file_group
                                                        ._v_name]))
                                     for net_file_group in h5f.root]))
                       for label, h5f in h5fs.iteritems()])
    
    # Generate flattened annealer results table, so we can apply
    # pandas manipulation.  The flattened table is in the form:
    #
    #     |    label   | namebase |  critical_path_delay  |  ...  |   width_fac  |
    #     |------------|----------|-----------------------|-------|--------------|
    #     |  vpr-fast  |   alu4   |       117.121n        |  ...  |       11     |
    #     |  vpr-fast  |   alu4   |       118.630n        |  ...  |       11     |
    flattened_results = flatten_dict_to_dataframe(OrderedDict([
            (k, flatten_dict_to_dataframe(v, 'namebase'))
            for k, v in data.iteritems()]), label_name)
    return flattened_results
```

```python
def get_max_min_channel_widths(h5fs):
    flattened_results = get_flattened_route_states(h5fs)
    max_min_channel_widths = (pd.pivot_table(flattened_results,
                                             values=['width_fac'],
                                             rows=['label',
                                                   'namebase'],
                                             aggfunc=np.max)
                              .transpose().rename(
                                      {'width_fac':
                                       'max_min_channel_widths'}))
    flattened_max_min = pd.melt(max_min_channel_widths)
    return pd.pivot_table(flattened_max_min,
                          values=['value'],
                          cols=['namebase'], aggfunc=np.max)
```

```python
def get_routing_data_frame(hdf_route_states, dtype=None):
    data = np.array([v.fetch_all_fields()
                     for v in hdf_route_states],
                    dtype=dtype)
    return pd.DataFrame(data)

    
def get_routes_to_do(h5fs, channel_widths_by_namebase):    
    combined = {}
    
    for label, h5f in h5fs.iteritems():
        results = {}
        for net_file_group in h5f.root:
            net_file_nambase = net_file_group._v_name
            hdf_route_states = net_file_group.route_states
            
            all_sha1s = set(hdf_route_states.cols
                            .block_positions_sha1[:])
            target_width_fac = (channel_widths_by_namebase
                                [net_file_nambase])
            data = get_routing_data_frame(hdf_route_states
                                          .where('(width_fac  == '
                                                 '%d)' %
                                                 target_width_fac),
#                                           .where('(success == True) & '
#                                                  '(width_fac  == %d)'
#                                                  % target_width_fac),
                                          dtype=hdf_route_states.dtype)
            if len(data) > 0:
                success_sha1s =  set(data['block_positions_sha1'])
            else:
                success_sha1s = set()
            results[net_file_nambase] = (target_width_fac,
                                         (all_sha1s - success_sha1s))
        combined[label] = results
    combined.keys()
    return combined
```

```python
from pprint import pprint
from path import path

def get_route_commands_for_channel_widths(combined):
    # sqsub -r 90m -o routed-cyplace-logs/route-${sha1}.log
    # --verbose --mpp 4G
    # ~/run_vpr_route_from_hdf.sh src/synced_results/
    #  placed-results-cyplace/placed-results-cyplace-7a8d71d4.h5 ${sha1}
    commands = []
    for label, results in combined.iteritems():
        for namebase, (width_fac, to_do) in results.iteritems():
            for sha1 in v:
                sha1_label = '%s-w%d' % (sha1, width_fac)
                output_path = (path('routed-%s-logs' % label)
                               .joinpath('route-%s.log' % sha1_label))
                command = ('mkdir -p routed-%s-logs ; '
                           'sqsub -r 90m -o %s --verbose --mpp 4G '
                           '~/run_vpr_route_from_hdf.sh '
                           'src/synced_results/placed-%s.h5 '
                           '%s src/synced_results/routed-%s %s' %
                           (label, output_path, label, sha1, label,
                            width_fac[0]))
                commands.append((label, namebase, sha1, 'w%d' %
                                 width_fac[0], command))
    return commands
```

```python
import cStringIO as StringIO


def get_routes_to_do_script(h5fs, channel_widths_by_namebase):
    output = StringIO.StringIO()
    print >> output, '#!/bin/bash'
    combined = get_routes_to_do(h5fs, channel_widths_by_namebase)
    commands = get_route_commands_for_channel_widths(combined)
    for command in commands:
        print >> output, '# %s' % (command[:-1], )
        print >> output, command[-1]
    return output.getvalue()
```

The figure below shows the _critical-path-delay_ and
_minimum-routable-channel-width_ for all trials of all three annealers
described above.  In each plot, the box extends from the lower to
upper quartile values of the data, with a line at the median.
The whiskers extend from the box to show the range of the
data.  The string [`7a8d71d4`][1] in the label [`cyplace 7a8d71d4`][1]
refers to the specific revision of the [`cyplace` _git_ repository][2] that
the [`cyplace`][2] trials were run with.

To reiterate, in each plot, the first box corresponds to running VPR's
_bounding-box_ place algorithm with `inner_num` set to 10, which is the
default.  The second box corresponds to running VPR's
_bounding-box_ place algorithm in _fast_ mode, which sets `inner_num` to one.
The `inner_num` parameter controls the number of moves attempted at each
temperature value, before the temperature schedule is updated for the next
batch of moves.  The number of moves attempted at each temperature is equal
to
$\mathtt{inner\_num} \times \left(\mathtt{block\_count}^{1.3333}\right)$, where $\mathtt{block\_count}$ corresponds to the nubmer of blocks in the
netlist.

[1]: https://github.com/gwg-placement-ga/cyplace/commit/7a8d71d4617d7f063006e18457cc6d9248b721d8
[2]: https://github.com/gwg-placement-ga/cyplace

From the plots in the figure below, we can see that in _most_ cases the
`cyplace`-annealer matches or out-performs VPR running in _`fast`_ mode.
_N.B._, the `cyplace`-annealer and the "VPR `fast`-annealer" are the two
right-most boxes in each plot. Perhaps more interesting is that there are
some cases where the `cyplace`-annealer seems to be performing in the same
ball-park as VPR running _not_ in `fast` mode, _i.e._, with `inner_num=10`.

Note that these are simply initial observations.  Things that still need to
be done, but are planned:

  * Compare _critical-path-delay_ for each netlist from the three annealers
    using the maximum routable _minimum-channel-width_ for any of the trials
    for the corresponding netlist.
      * This should give a fairer comparison, since relaxing the routing
        channel-width constraints may help reduce the _critical-path-delay_
        for some placements.
  * Perform pairwise statistical comparison of _minimum-channel-width_ and
    _critical-path-delay_ between the three annealers, to obtain $p$-values.
  * Compare aggregated stats, grouped by annealer, _e.g._, geo-mean, sum of
    mean _critical-path-delay_ and mean _minimum-channel-width_, for all
    trials.
