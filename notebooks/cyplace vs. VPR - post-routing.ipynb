In this section, we discuss the post-routing results of placing all 20 MCNC
netlist benchmarks using:

 1. VPR's place bounding-box annealer,
    running in _fast_ mode _(i.e., `inner_num=1`)_
 2.  VPR's bounding-box place annealer,
    running in default _fast_ mode _(i.e., `inner_num=10`)_
 3. `cyplace` annealer using _Star+_
 
For each annealer, we conducted 10 trials, _i.e._, we placed each netlist 10
times with 10 unique random-number-generator seeds.

```python
from collections import OrderedDict

import pandas as pd
import tables as ts
from cyvpr.result.routing_pandas import get_routing_data_frame, min_success_data

pd.set_eng_float_format(accuracy=3, use_eng_prefix=True)
pylab.rcParams['figure.figsize'] = (10.0, 4)
```

```python
h5f_paths = OrderedDict([
        ('vpr-not_fast-bounding box',
         '../../synced_results/routed-not_fast-bounding_box.h5'),
        ('vpr-fast-bounding box',
         '../../synced_results/routed-fast-bounding_box.h5'),
        ('cyplace-7a8d71d4',
         '../../synced_results/routed-cyplace-7a8d71d4.h5'),
])
```

```python
h5fs = OrderedDict([(k, ts.open_file(v, 'r'))
                    for k, v in h5f_paths.iteritems()])
```

# Plot VPR-anneal vs. cyplace-anneal post-routing results

```python
data = OrderedDict([(label,
                    OrderedDict([(net_file_group._v_name,
                                  min_success_data(
                                   get_routing_data_frame(
                                    net_file_group.route_states)))
                                 for net_file_group in h5f.root]))
                   for label, h5f in h5fs.iteritems()])
```

The figure below shows the _critical-path-delay_ and
_minimum-routable-channel-width_ for all trials of all three annealers
described above.  In each plot, the box extends from the lower to
upper quartile values of the data, with a line at the median.
The whiskers extend from the box to show the range of the
data.  The string [`7a8d71d4`][1] in the label [`cyplace 7a8d71d4`][1]
refers to the specific revision of the [`cyplace` _git_ repository][2] that
the [`cyplace`][2] trials were run with.

To reiterate, in each plot, the first box corresponds to running VPR's
_bounding-box_ place algorithm with `inner_num` set to 10, which is the
default.  The second box corresponds to running VPR's
_bounding-box_ place algorithm in _fast_ mode, which sets `inner_num` to one.
The `inner_num` parameter controls the number of moves attempted at each
temperature value, before the temperature schedule is updated for the next
batch of moves.  The number of moves attempted at each temperature is equal
to
$\mathtt{inner\_num} \times \left(\mathtt{block\_count}^{1.3333}\right)$, where $\mathtt{block\_count}$ corresponds to the nubmer of blocks in the
netlist.

[1]: https://github.com/gwg-placement-ga/cyplace/commit/7a8d71d4617d7f063006e18457cc6d9248b721d8
[2]: https://github.com/gwg-placement-ga/cyplace

From the plots in the figure below, we can see that in _most_ cases the
`cyplace`-annealer matches or out-performs VPR running in _`fast`_ mode.
_N.B._, the `cyplace`-annealer and the "VPR `fast`-annealer" are the two
right-most boxes in each plot. Perhaps more interesting is that there are
some cases where the `cyplace`-annealer seems to be performing in the same
ball-park as VPR running _not_ in `fast` mode, _i.e._, with `inner_num=10`.

Note that these are simply initial observations.  Things that still need to
be done, but are planned:

  * Compare _critical-path-delay_ for each netlist from the three annealers
    using the maximum routable _minimum-channel-width_ for any of the trials
    for the corresponding netlist.
      * This should give a fairer comparison, since relaxing the routing
        channel-width constraints may help reduce the _critical-path-delay_
        for some placements.
  * Perform pairwise statistical comparison of _minimum-channel-width_ and
    _critical-path-delay_ between the three annealers, to obtain $p$-values.
  * Compare aggregated stats, grouped by annealer, _e.g._, geo-mean, sum of
    mean _critical-path-delay_ and mean _minimum-channel-width_, for all
    trials.

```python
net_file_namebases = sorted(data.values()[0].keys())

fig = figure(figsize=(10, 6 * len(net_file_namebases)))
fig.subplots_adjust(hspace=0.5)

for i, net_file_namebase in enumerate(net_file_namebases):
    fields = ('critical_path_delay', 'width_fac')
    for j, field in enumerate(fields):
        subplot_index = (len(net_file_namebases), len(fields),
                         i * len(fields) + j + 1)
        ax = fig.add_subplot(*subplot_index)
        box_plot = ax.boxplot([d[net_file_namebase][field]
                 for d in data.itervalues()])
        ax.set_title('[%s] %s' % (net_file_namebase, field))
        ax.set_xticklabels([' '.join(k.split('-')[:2]) for k in data.keys()])
```
