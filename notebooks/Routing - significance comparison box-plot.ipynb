```python
import itertools
from collections import OrderedDict

import tables as ts
import pandas as pd
from cyvpr.result.routing_pandas import (get_max_min_channel_widths,
                                          get_routing_data_frames)
from cyvpr.bin.compare_routed import (
        boxplot_route_state_frames,
        route_data_frames_significance_comparison)

pd.set_eng_float_format(accuracy=3, use_eng_prefix=True)
pylab.rcParams['figure.figsize'] = (10.0, 4)
# pylab.rcParams['font.size'] = 16

def p_value_from_labels(p_values_frame, a_label, b_label):
    # Given two labels, look-up p-value from table
    row = p_values_frame[(p_values_frame['A'] == a_label)
                          & (p_values_frame['B'] == b_label)]
    if not row:
        row = p_values_frame[(p_values_frame['A'] == b_label)
                              & (p_values_frame['B'] == a_label)]
    if not row:
        raise KeyError, ('No p-value row matching labels: %s and %s'
                         % (a_label, b_label))
    return row
```

```python
labels = 'vpr-not_fast-bounding_box vpr-fast-bounding_box cyplace-7a8d71d4'.split(' ')
h5f_paths = OrderedDict([(label, '../../sharcnet_place_management/'
                          'routed-%s.h5' % label)
                         for label in labels])
h5fs = OrderedDict([(k, ts.open_file(v, 'r'))
                    for k, v in h5f_paths.iteritems()])
max_min_channel_widths = get_max_min_channel_widths(h5fs)
data_frames = get_routing_data_frames(h5fs, max_min_channel_widths)
```

```python
def significance_boxplot_route_state_frames(data_frames,
                                            net_file_namebase,
                                            column):
    fig, ax, data_vectors = boxplot_route_state_frames(
            data_frames, net_file_namebase, column)
    data_vectors = np.array(data_vectors)
    labels = np.array(data_frames.keys())
    colors = ['red', 'green']
    compare_results = (
            route_data_frames_significance_comparison(data_frames,
                                                      net_file_namebase,
                                                      column))
    
    yticks = ax.get_yticks()
    ytick_extent = yticks[1] - yticks[0]
    pairs = np.array(list(itertools
                          .combinations(range(len(data_frames)),
                                        2)))
    
    for pair in pairs:
        vectors = data_vectors[pair]
        p_value_row = p_value_from_labels(compare_results,
                                          *labels[pair])
        p_value = p_value_row['p-value']
        means = np.array([v.mean() for v in vectors])
        mean_diff = means[1] - means[0]
        ax.annotate('', xy=((pair[0] + 1) + .15, means[0]),
                    xytext=((pair[1] + 1) - .15, means[1]),
                    arrowprops={'arrowstyle': '<->',
                                'color':
                                colors[int(p_value < 0.05)],
                                'alpha': 0.7,
                                'linewidth': 3})
        x_center = (pair[1] + pair[0] + 2) / 2.
        if (pair[1] - pair[0]) % 2:
            y_factor = 1
        else:
            y_factor = -1
        ax.annotate('p-value: %.2g' % p_value,
                    xy=(x_center, means.mean()),
                    ha='center',
                    fontsize=16,
                    xytext=(x_center, y_factor * 0.65
                            * ytick_extent + means.mean()),
                    bbox=dict(boxstyle='round', fc='white',
                              alpha=0.9))
    return fig, ax, data_vectors, compare_results
```

# Plots

Below, we plot the routing critical-path-delay results from running 10 trials
for each of the following place algorithms:

 * `vpr-not_fast-bounding_box`
 * `vpr-fast-bounding_box`
 * `cyplace-7a8d71d4`

The routings for each netlist were all performed with the same channel-width,
which was chosen as the maximum channel-width that all placer algorithms
could successfully route the placements from all 10 placement trials of the
corresponding netlist.

Note the _$p$-value_ significance markings on the plots.  The significance
_$p$-values_ are computed using a [Wilcoxon signed-rank test][1] and are
labelled on the plot above either a _red_ or _green_ arrow, pointing to the
relevant means in the box plot.

A _green_ arrow indicates that the $p$-value is less than 0.05, such that the
null-hypothesis may be rejected, indicating that the means are not
statistically likely to be due to drawing from the same population.  If the
$p$-value is greater than 0.05, the arrow is _red_.

[1]: http://en.wikipedia.org/wiki/Wilcoxon_signed-rank_test

## VPR _not-fast_ vs VPR _fast_ vs `cyplace`

In the figures below, we plot VPR in both the default bounding-box mode
_(i.e., `inner_num=10`)_ and in _fast_ mode _(i.e., `inner_num=1`)_, as well
as our proposed annealer, which uses Star+, called `cyplace`.  In the
majority of cases, we observe that there is either no significant difference
in the results from `cyplace` compared to VPR run with `inner_num=10`.  In
fact, there are only 3 cases where VPR with `inner_num=10` out-performs
_(i.e., achieves a lower critical-path-delay)_ with statistical significance.

```python
net_file_namebases = sorted(data_frames.values()[0].keys())

for net_file_namebase in net_file_namebases:
    significance_boxplot_route_state_frames(data_frames,
                                            net_file_namebase,
                                            'critical_path_delay')
```

## VPR _fast_ vs `cyplace`

In the figures below, we plot only VPR in _fast_ mode _(i.e., `inner_num=1`)_
and our proposed annealer, which uses Star+, called `cyplace`.  Note that
there is only one netlist, `dsip`, where VPR achieves a mean
_critical-path-delay_ that is statistically significantly lower than
`cyplace`.

```python
net_file_namebases = sorted(data_frames.values()[0].keys())

for net_file_namebase in net_file_namebases:
    significance_boxplot_route_state_frames(OrderedDict(data_frames
                                                        .items()
                                                        [1:]),
                                            net_file_namebase,
                                            'critical_path_delay')
```
